# `__str__ and __repr__`

```python
class Student:
  college = "HCL GUVI"

  def __init__(self,name):
    self.name = name

  def __str__(self):
    return f"Student {self.name}'s profile has been successfully created"


  def greeting(self):
    print(f"Hello {self.name}, Welcome to {self.college}")
```

```python
s1 = Student("Akash")
```

```python
print(s1)
```

```python
class Student:
  college = "HCL GUVI"

  def __init__(self,name):
    self.name = name

  def __str__(self):
    return f"Student {self.name}'s profile has been successfully created"

  def __repr__(self):
    return f"Hello {self.name}, Welcome to {self.college}"


  def greeting(self):
    print(f"Hello {self.name}, Welcome to {self.college}")
```

```python
s3 = Student("Kabilnath")
```

```python
s5 = Student("Sreehari")
```

```python
print([s3,s5])
```

```python
print(s3)
print(repr(s3))
```

```python
class Student:
  college = "HCL GUVI"

  def __init__(self,name):
    self.name = name

  def __repr__(self):
    return f"Hello {self.name}, Welcome to {self.college}"


  def greeting(self):
    print(f"Hello {self.name}, Welcome to {self.college}")
```

```python
s4 = Student("Deepa Lakshmi")
print(s4)
```

if you are using direct print, first it will prioritize `__str__`, if there is no str then it will automatically fall back tp `__repr__`

| Statement                        | Method Called                                          |
| -------------------------------- | ------------------------------------------------------ |
| `print(obj)`                     | `__str__()` (or `__repr__()` if `__str__()` is absent) |
| `str(obj)`                       | `__str__()`                                            |
| `repr(obj)`                      | `__repr__()`                                           |
| Typing `obj` in the Python shell | `__repr__()`                                           |
| Elements displayed in a list     | `__repr__()`                                           |


### Difference Betwen `__str__() and __repr__()`

| Feature     | `__str__()`                          | `__repr__()`                                        |
| ----------- | ------------------------------------ | --------------------------------------------------- |
| Purpose     | User-friendly display                | Developer/debugging display                         |
| Called By   | `print()`, `str()`                   | `repr()`, interactive shell, containers like `list` |
| Readability | Simple and readable                  | Detailed and unambiguous                            |
| Fallback    | Falls back to `__repr__()` if absent | Used directly if called                             |

### Static Method

```python
class Banking:

  @staticmethod
  def security_check():
    print("Welcome to the bank, please go inside")
```

```python
c1 = Banking()
```

```python
c1.security_check()
```

### Class & Instance method

```python
class Banking:
  name = "SBI Bank"

  @classmethod
  def show_details(cls):
    print(f"Welcome to {cls.name}")

  @classmethod
  def modify_name(cls,new_name):
    cls.name = new_name
    print(f"The name of the bank has been changed to {cls.name}")

  def __init__(self,name):
    self.name = name
    print(f"Welcome to the Bank, {self.name}")

  def modify_customer_detaile(self,new_name):
    self.name = new_name
    print(f"The name of the customer has been changed to {self.name}")
```

```python
c2 = Banking("Durga")
```

```python
c2.modify_customer_detaile("Durga Priyadharshini")
```

```python
c2.show_details()
```

```python
c2.modify_name("Indian Bank")
```

```python
c3 = Banking("Jaidiksha")
```

```python
c3.show_details()
```

| Feature                    | Instance Method    | Class Method                | Static Method                                     |
| -------------------------- | ------------------ | --------------------------- | ------------------------------------------------- |
| Decorator                  | None               | `@classmethod`              | `@staticmethod`                                   |
| First Parameter            | `self`             | `cls`                       | None                                              |
| Access Instance Attributes | ✅ Yes              | ❌ No                        | ❌ No                                              |
| Access Class Attributes    | ✅ Yes              | ✅ Yes                       | ❌ No (unless referenced explicitly by class name) |
| Modify Instance Attributes | ✅ Yes              | ❌ No                        | ❌ No                                              |
| Modify Class Attributes    | ✅ Yes              | ✅ Yes                       | ❌ No (unless referenced explicitly by class name) |
| Called Using               | Object (preferred) | Class (preferred) or Object | Class (preferred) or Object                       |


## Inheritance

### Simple/Single Inheritance

```python
class RBI:

  @staticmethod
  def interest_rate():
    print("Interest rate is 9% for Housing Loan")
    print("Interest rate is 12% for Personal Loan")
    print("Interest rate is 15% for Car Loan")
```

```python
class SBI(RBI):

  def loan(self):
    print("SBI provides Housing Loan")
    print("SBI provides Personal Loan")
    print("SBI provides Car Loan")
```

```python
c1 = SBI()
```

```python
c1.loan()
```

```python
c1.interest_rate()
```

```python
rbi_c1 = RBI()
rbi_c1.interest_rate()
```

```python
rbi_c1.loan()
```

### Multiple Inheritance

```python
class father():

  def __init__(self):
    print("I am a father-parent class")

  def father_assests(self):
    print("I have a house")
    print("I have 2 cars")
    print("I have 5 commerical buildings")

class mother():

  def __init__(self):
    print("I am a mother-parent class")

  def mother_assests(self):
    print("I have a house")
    print("I have a car")
    print("I have a 1000g of gold")


class son(father,mother):

  def __init__(self):
    print("I am a son class")

  def son_assests(self):
    print("I have a car and a super bike")
```

```python
child1 = son()
```

```python
child1.son_assests()
```

```python
child1.father_assests()
```

```python
child1.mother_assests()
```

### Multi-level inheritance

```python
class Grand_father():

  def __init__(self):
    print("I am a Grand_father-class")

  def grandfather_assests(self):
    print("I have a house")
    print("I have 100 acres of farm land")

class father(Grand_father):

  def __init__(self):
    print("I am a father-class")

  def father_assests(self):
    print("I have 5 commerical buildings")
    print("I am founder of a MNC")


class son(father):

  def __init__(self):
    print("I am a son class")

  def son_assests(self):
    print("I have a car and a super bike")
```

```python
child2 = son()
```

```python
child2.son_assests()
```

```python
child2.father_assests()
```

```python
child2.grandfather_assests()
```

### Hierarchical Inheritance

```python
class College:

  college_name = "HCL GUVI"

  def login(self):
    print("your login has been successful captured")

  def logout(self):
    print("your login has been successful captured")
```

```python
class student(College):
  def __init__(self,name):
    self.name = name

  def profile(self):
    print(f"Name: {self.name}")

  def subject(self):
    print("Your degree covers the following subjects")
    print("Python")
    print("SQL")
    print("Data Analysis")
    print("Data Visualization")

  def project(self):
    print("You have to complete 5 mini projects and 1 major project")
```

```python
s1 = student("Deepa Lakshmi")
```

```python
s1.profile()
```

```python
s1.subject()
```

```python
s1.project()
```

```python
class teaching_staff(College):
  def __init__(self,name):
    self.name = name

  def add_profile(self,*subjects):
    print(f"Name: {self.name}, can able to teach below listed subjects")
    self.subjects = subjects

  def display_profile(self):
    print(f"Name: {self.name}, can able to teach below listed subjects")
    for subject in self.subjects:
      print(subject)

  def experience(self, year_of_experience):
    print(f"You have {year_of_experience} years of experience")
```

```python
t1 = teaching_staff("Sreehari")
```

```python
t1.add_profile("Python","SQL","Data Analysis","Data Visualization")
```

```python
t1.display_profile()
```

```python
t1.experience(5)
```

```python
t2 = teaching_staff("Kabilnath")
```

```python
t2.login()
```

```python
s3 = student("Akash")
```

```python
s3.login()
```

```python
t2.logout()
```

```python
s3.logout()
```
