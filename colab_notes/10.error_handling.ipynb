```python
try:
  number1 = int(input())
  number2 = int(input())
  print(number1/number2)
except ZeroDivisionError:
  print(f"Value of Number2  is 0, we can't perform division with 0")
except ValueError:
  print("Invalid Input, please re-try with numerical input")
except Exception as e:
  print(f"Invalid Input: {e}")
else:
  print("Program executed successfully without any error")
finally:
  print("program ended")
```

```python
try:
  number1 = int(input("Enter your pin number"))
except ValueError:
  print("Invalid Input, please re-try with numerical input")
else:
  if number1 == 1234:
    print("Welcome to your account")
    account_balan = 10000
    withdraw = int(input("Enter the amount you want to withdraw"))
    if withdraw > account_balan:
      print("Insufficient balance")
    else:
      account_balan = account_balan-withdraw
      print(f"Your account balance is {account_balan}")
  else:
    print("Invalid pin")
finally:
  print("Thank you, Visit again")
```

```python
class insufficientBalanceError(Exception):
  pass

try:
  number1 = int(input("Enter your pin number"))
except ValueError:
  print("Invalid Input, please re-try with numerical input")
else:
  if number1 == 1234:
    print("Welcome to your account")
    account_balan = 10000
    withdraw = int(input("Enter the amount you want to withdraw"))
    try:
      if withdraw > account_balan:
        raise insufficientBalanceError("Insufficient balance")
      else:
        account_balan = account_balan-withdraw
        print(f"Your account balance is {account_balan}")
    except insufficientBalanceError as e:
      print(e)
  else:
    print("Invalid pin")
finally:
  print("Thank you, Visit again")
```
