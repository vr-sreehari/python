# Conditional Statement - Cont.

* if
* elif
* else

```python
#200, 500, 750, 1000, 1000+
#Free, 5,   7,    14,  20

consumed_units = int(input("Enter the number of units consumed: "))
if consumed_units <= 1000:
    if consumed_units <= 200:
        print("Your bill amount is: Free")
    elif consumed_units <= 500:
        unit_split = consumed_units - 200
        bill_amount = unit_split * 5
        print("Your bill amount is:", bill_amount)
    elif consumed_units <= 750:
        unit_split = consumed_units - 200
        split1 = unit_split - 300
        bill_amount = (split1 * 7) + (300 * 5)
        print("Your bill amount is:", bill_amount)
    else:
        unit_split = consumed_units - 200
        split = unit_split - 300 - 250
        bill_amount = (split * 14) + (300 * 5) + (250 * 7)
        print("Your bill amount is:", bill_amount)

else:
    bill_amount = consumed_units*20
    print(bill_amount)
```

# **Looping Statement**

* Multiple Iteration
* two types

## **while**

image.png

```python
Rank = 1
while Rank <= 10:
  print(Rank)
  Rank += 1
print("End of Loop")
```

```python
attempt = 0
while attempt < 3:
  print("Attempt #", attempt," Over" )
  attempt += 1
print("Try after sometime")
```

```python
ticket = 5
while ticket > 0:
  print("You are allowed to enter")
  ticket -= 1
print("No more tickets")
```

```python
N = int(input())
i = 1
while i <= N:
  if i%2 == 0:
    print(i)
  i += 1
```

### additional options for iterables

* range()
* enumerate()
* zip()

```python
tickets = int(input())
for i in range(tickets):
  print(i)
```

```python
student = ["Kokil","Vishnu","Lokesh","Durga","Agalya","Rahul"]
for index, name in enumerate(student):
  print("Stunder number #", index+1, "is ", name)
```

```python
student_name = ["Kokil","Vishnu","Lokesh","Durga","Agalya","Rahul"]
student_mark = [95,98,94,97,99,96]

for name,mark in zip(student_name,student_mark):
  print(name, ":", mark)
```

```python
for i in range(len(student_name)):
  name = student_name[i]
  mark = student_mark[i]
  print(name, ":", mark)
```

```python
for i in range(10):
  for j in range(5):
    print(i, "*",j, "=", i*j)
  print("\n")
  print("---"*10)
```

## Loop Control Statement

* **break** → "Stop the entire loop."
* **continue** → "Skip this iteration and keep looping."
* **pass** → "Do nothing at all—it's just a placeholder."

## Break

```python
for i in range(1,50):
  if i%5 == 0:
    break
  print(i)
```

## Continue

```python
for i in range(1,50):
  if i%5 == 0:
    continue
  print(i)
  print("--")
```

## Pass

```python
bill = int(input())
if bill < 1000:
  pass
else:
  print("I will pay my bill within 5")
```
