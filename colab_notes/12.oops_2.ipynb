### Method Overriding

```python
class shape:
  def area(self):
    print("Area of an object can be calculated based on it's shape")

class square(shape):
  def area(self):
    print("Area of a square: Lenght**2 ")

class rectangle(square):
  def area(self):
    print("Area of a rectangle: (Lenght * Breadth)")
```

```python
rect1 = rectangle()
```

```python
rect1.area()
```

```python
class shape:
  def area(self):
    print("Area of an object can be calculated based on it's shape")

class square(shape):
  def area(self):
    print("Area of a square: Lenght**2 ")
    super().area()

class rectangle(square):
  def area(self):
    print("Area of a rectangle: (Lenght * Breadth)")
    super().area()
```

```python
rect2 = rectangle()
```

```python
rect2.area()
```

### Method override in multiple inheritance

```python
class father():

  def __init__(self):
    print("I am a father-parent class")

  def assests(self):
    print("I have a house")
    print("I have 2 cars")
    print("I have 5 commerical buildings")

class mother():

  def __init__(self):
    print("I am a mother-parent class")

  def assests(self):
    print("I have a house")
    print("I have a car")
    print("I have a 1000g of gold")


class son(father,mother):

  def __init__(self):
    print("I am a son class")

  def assests(self):
    print("I have a car and a super bike")
```

```python
son1 = son()
son1.assests()
```

```python
class father():

  def __init__(self):
    print("I am a father-parent class")

  def assests(self):
    print("--"*10)
    print("I have a house")
    print("I have 2 cars")
    print("I have 5 commerical buildings")
    super().assests()

class mother():

  def __init__(self):
    print("I am a mother-parent class")

  def assests(self):
    print("--"*10)
    print("I have a house")
    print("I have a car")
    print("I have a 1000g of gold")


class son(father,mother):

  def __init__(self):
    print("I am a son class")

  def assests(self):
    print("--"*10)
    print("I have a car and a super bike")
    super().assests()
```

```python
son1 = son()
son1.assests()
```

## Abstraction

Scenario: Working of ATM Machine

```python
from abc import ABC,abstractmethod
```

```python
class ATM_Machine_demo(ABC):

  @abstractmethod
  def login(self):
    print("This method will verify the user credential")

  @abstractmethod
  def check_balance(self):
    print("This method will call login method internally and then print current balance of a customer")

  @abstractmethod
  def withdraw(self,amount):
    pass

  @abstractmethod
  def deposit(self,amount):
    pass
```

```python
ca1 = ATM_Machine_demo()
```

```python
class ATM_Machine(ATM_Machine_demo):
  customer = {"Kaviya_123":{"pin":4578,"balance":10000},
              "Captian_Miller":{"pin":7896,"balance":5000},
              "Vivek":{"pin":1234,"balance":100000}}
  def __init__(self,name,pin):
    print("Welcome to ATM Machine")
    self.name = name
    self.pin = pin

  def login(self):
    if self.name in self.customer and self.customer[self.name]["pin"] == self.pin:
      print("Login successful")
      return True
    else:
      print("Login Failed")
      return False

  def check_balance(self):
    if self.login():
      print("Available balance: ",self.customer[self.name]["balance"])

  def withdraw(self,amount):
    if self.login():
      if self.customer[self.name]["balance"] >= amount:
        self.customer[self.name]["balance"] -= amount
        print("Available balance: ",self.customer[self.name]["balance"])
      else:
        print("Insufficient balance")

  def deposit(self,amount):
    if self.login():
      self.customer[self.name]["balance"] += amount
      print("Available balance: ",self.customer[self.name]["balance"])

```

```python
cust1 = ATM_Machine("Kaviya_123",4578)
```

```python
cust1.check_balance()
```

```python
cust1.withdraw(5000)
```

```python
cust2 = ATM_Machine("Captian_Miller",7896)
```

```python
cust2.deposit(10000)
```

* Protected Variable -> (_Numvar) -> we should access only inside a class or sub-class
