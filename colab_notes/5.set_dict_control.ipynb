## Real-Time Use Cases

### Remove Duplicates in a list

```python
a = [0,1,2,3,4,5,6,1,2,5,8,3]
set_a = set(a)
print(set_a)
```

### Common elements between two lists

```python
a = [1,2,3,4,5]
b = [4,5,6,7,8]
set_a = set(a)
set_b = set(b)
print(set_a.intersection(set_b))
```

```python
family_list = [("Male",30),("Female",25),("Child1",10),("Child2",5)]
for i in family_list:
  #print(i)
  if i[0] == "Male":
    print("Full Ticket")
  elif i[0] == "Female":
    print("Free Ticket")
  else:
    if i[1] <= 7:
      print("Free Ticket")
    else:
      print("Full Ticket")
  print("------")
```

```python
family_list = [("Male",30),("Female",25),("Child1",10),("Child2",5)]
for i in family_list:
  #print(i)
  if i[0] == "Male":
    print(i[0]," : ","Full Ticket")
  elif i[0] == "Female":
    print(i[0]," : ""Free Ticket")
  else:
    if i[1] <= 7:
      print(i[0]," : ""Free Ticket")
    else:
      print(i[0]," : ""Full Ticket")
  print("------")
```
