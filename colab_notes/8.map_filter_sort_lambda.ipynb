## Filter function using `None`

## `None` as expression in Filter

Why does None work?

When None is passed as the first argument, filter() removes all falsy values.

Falsy values include:

0

""

[]

{}

False

None

```python
raw_list = [1,2,3,0,"Arun",False,"GUVI",[],{},None]
cleaned_list = list(filter(None,raw_list))
```

```python
cleaned_list
```

```python
raw_list = [0,1,2,3,"Arun",False,"GUVI",[],{},None]
cleaned_list = list(filter(None,raw_list))
cleaned_list
```

```python
raw_list = [5.0,1,2,3,"Arun",False,"GUVI",[],{},None]
cleaned_list = list(filter(None,raw_list))
cleaned_list
```

```python
def falsy(x):
  if x == "":
    return False
  elif x == None:
    return False
  elif x == {}:
    return False
  else:
    return True
```

```python
raw_list = [5.0,1,2,3,0,"Arun",False,"GUVI",[],{},None]
cleaned_list = list(filter(falsy,raw_list))
cleaned_list
```

```python
a = [0,1,2,3,4,[],["",0]]
a[6] = list(filter(None,a[6]))
a
```

```python
list(filter(None,a))
```

## Sort using user-defined function

```python
def sort_mark(student):
  return student[1]

student_marks = {
  "Arun": 90,
  "Vinoth": 80,
  "Akash":65,
  "Selva":75}

sorted(student_marks.items(), key=sort_mark)
```

# Lambda Function

Usage
* It's user-defined function
* Single line function
* For Simple task
* No need to use `def' keyword to define a func,
* Use `Lambda`

## Without Arguments

Normal Function Vs Lambda

```python
def Greeting():
  print("Hello All, Good Morning")

Greeting()
```

### Syntax

func_name = `Lambda: Expression`

```python
greet = lambda: "Hello All, Good Morning"
print(greet())
```

```python
greet = lambda: """Hello All,
Good Morning
will start our session shortly"""
print(greet())
```

## With Arguments

### Syntax

`func_name = lambda args: expression`

```python
def Greeting(name):
  print(f"Hello {name}, Good Morning")

Greeting("Arun")
```

```python
greet = lambda name: f"Hello {name}, Good Morning"
print(greet("Arun"))
```

```python
print(greet("Joe"))
```

```python
addition = lambda a,b: f"Sum of {a} and {b} is {a+b}"
addition(50,25)
```

## With Arguments and Conditions

```python
def even(a,b):
  total = (a+b)
  if total%2==0:
    print("even")
  else:
    print("odd")
```

```python
n = int(input())
m = int(input())
even(n,m)
```

* it allows only if and else, not elif

### Syntax

if and else:

`function_name = lambda arguments: "true_expression" if condition_1 else "False_expression"`

```python
even = lambda a: "Even" if a%2==0 else "Odd"
```

```python
print(even(15))
```

```python
print(even(14))
```

```python
odd_even = lambda a,b: "Even" if (a+b)%2==0 else "Odd"
```

```python
print(odd_even(14,15))
```

## with Map()

### Syntax

`list(map(expression,iterables))`<br>

syntax for expression:
`lambda arguments: function`

```python
#square = lambda x: x**2
b = [10,20,30,40,50]
squared_number = list(map(lambda x: x**2,b))
squared_number
```

## Lambda with Filter()

### Syntax

`list(filter(expression,iterables))`<br>

syntax for expression:
`lambda arguments: function`

```python
fruits = ["Apple","Banana","Cherry","Grape","Fig","Jackfruit"]
filtered_fruits = list(filter(lambda fruit_name: len(fruit_name)>5,fruits))
filtered_fruits
```

## Lambda with Sorted()

### Syntax

`sorted(iterables,key=expression,reverse = Boolean)`<br>

syntax for expression:
`lambda arguments: function`

```python
sorted(student_marks.items(),key=lambda student: student[1])
```

```python
sorted(student_marks.items(),key=lambda student: student[1],reverse=True)
```

| Lambda                    | Meaning                                             |
| ------------------------- | --------------------------------------------------- |
| `lambda: 100`             | No arguments                                        |
| `lambda x: x*x`           | One argument                                        |
| `lambda x, y: x+y`        | Two arguments                                       |
| `lambda *args: sum(args)` | Variable positional arguments                       |
| `lambda **kwargs: kwargs` | Variable keyword arguments (valid, though uncommon) |


```python
addition = lambda *args: sum(args)
```

```python
print(addition(10,20,30,40))
```
