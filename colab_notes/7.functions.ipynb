#Lifetime of local variable declared in control flow statement

Any variable declared inside control flow statement like if, for, and while statements are valid only inside the block

```python
#200, 500, 750, 1000, 1000+
#Free, 5,   7,    14,  20

consumed_units = int(input("Enter the number of units consumed: "))
if consumed_units <= 1000:
    if consumed_units <= 200:
        print("Your bill amount is: Free")
    elif consumed_units <= 500:
        unit_split = consumed_units - 200
        bill_amount = unit_split * 5
        print("Your bill amount is:", bill_amount)
    elif consumed_units <= 750:
        unit_split = consumed_units - 200
        split1 = unit_split - 300
        bill_amount = (split1 * 7) + (300 * 5)
        print("Your bill amount is:", bill_amount)
    else:
        unit_split = consumed_units - 200
        split = unit_split - 300 - 250
        bill_amount = (split * 14) + (300 * 5) + (250 * 7)
        print("Your bill amount is:", bill_amount)


else:
    bill_amount = consumed_units*20
    print(bill_amount)
print(split1)
```

## 5. **kwargs

```python
def student(**details):
  print(type(details))
  print(details)
```

```python
student(name="Jahidiksha",age=20,city="Chennai",state="Tamil Nadu",education="B.Tech")
```

```python
def student(**details):
  for i in details.items():
    print(i[0],":",i[1])
```

```python
student(name="Jahidiksha",age=20,city="Chennai",state="Tamil Nadu",education="B.Tech")
```

```python
student(name="Bernard",city="Bangalore",state="Karnataka",education="M.Tech")
```

## Using all in one example

```python
def student_info(name, age, *skill,year="I-Year",**details):
  print(name)
  print(age)
  print(year)
  for i in skill:
    print(i)
  for i in details.items():
    print(f"{i[0]} : {i[1]}")
```

## Global vs local variable

```python
def stock_selling(apple,banana,mango):
  func_total = (apple*30)+(banana*20)+(mango*50)
```

```python
stock_selling(10,20,5)
```

```python
func_total
```

```python
func_total = 0  #global variable
def stock_selling(apple,banana,mango):
  global func_total
  func_total = (apple*30)+(banana*20)+(mango*50) #local

stock_selling(10,20,5)
print(func_total)
```
