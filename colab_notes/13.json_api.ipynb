# API

It allows two different software programs to communicate and exchange data with each other with set of rules and protocols

API acts as a messenger between two applications.

APIs enable data exchange without exposing the internal implementation.


**Why Do We Need APIs?**
* Enable communication between applications.
* Share data securely.
* Avoid building everything from scratch.
* Integrate third-party services.
* Automate tasks and workflows.

## Request

**Request Contains:**

* URL
* HTTP Method (GET, POST...)
* Headers
* Parameters
* Body (optional)



**Which restaurant?** → URL <br>
**What do you want?** → Parameters <br>
**Who are you?** → Headers <br>
**What exactly should be prepared?** → Body <br>

### Methods

| Method | Purpose              |
| ------ | -------------------- |
| GET    | Retrieve data        |
| POST   | Create new data      |
| PUT    | Update existing data |
| DELETE | Remove data          |


```python
import requests
```

```python
response = requests.get("https://jsonplaceholder.typicode.com/")
```

### Headers

| Header        | Purpose                             |
| ------------- | ----------------------------------- |
| Authorization | Who is making the request?          |
| Content-Type  | What format is the data?            |
| Accept        | What response format do I want?     |
| User-Agent    | Which application sent the request? |


```python
url = "https://jsonplaceholder.typicode.com/"

header = {"Content-Type": "application/json",
          "Accept": "application/json"}

response = requests.get(url, headers=header)
```

### Parameters

Parameters are used to filter or customize the data you want.

They are appended to the URL.

```python
url = "https://api.agify.io"

parameter = {"name":"ganesh","country_id":"IN"}

response = requests.get(url,params=parameter)
response = requests.get("https://api.agify.io",params={"name":"ganesh","country_id":"IN"})
```

```python
response.text
```

## Response

**Response contains**

* Status Code
* Header
* Response data

### Status code

| Code    | Meaning               | Description                                                             |
| ------- | --------------------- | ----------------------------------------------------------------------- |
| **200** | OK                    | Request was successful.                                                 |
| **201** | Created               | A new resource was successfully created.                                |
| **400** | Bad Request           | The request sent by the client is invalid.                              |
| **401** | Unauthorized          | Authentication is required or invalid credentials were provided.        |
| **403** | Forbidden             | You are authenticated but don't have permission to access the resource. |
| **404** | Not Found             | The requested resource or endpoint doesn't exist.                       |
| **500** | Internal Server Error | Something went wrong on the server side.                                |
